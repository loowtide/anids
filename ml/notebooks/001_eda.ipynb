```python
from pathlib import Path
raw_dir=Path("../data/clean")
all_cols={}

for path in raw_dir.glob("*.csv"):
    df=pd.read_csv(path,nrows=1)
    all_cols[path.name]=set(df.columns.str.strip())

all_cols_union=set().union(*all_cols.values())
for name ,cols in all_cols.items():
    missing=all_cols_union-cols
    if missing:
        print(f"{name} is missing")
        print(missing)
```

```python
import re
```

```python
def normalize(col):
    col=col.strip()
    col=re.sub(r'[\s/\-]+','_',col)
    col=re.sub(r'[^\w]','',col)
    col=re.sub(r'_+','_',col)
    return col.lower().strip('_')

norm1={normalize(c): c for c in df1}
norm2={normalize(c): c for c in df2}
print(set(norm2))

```

```python
input_dir = Path("../data/clean")

files=input_dir.glob("*.csv")
col_sets={}

for f in files:
    cols=pd.read_csv(f,nrows=0).columns.tolist()
    col_sets[f]=set(cols)

all_cols=set()
for cols in col_sets.values():
    all_cols |=cols

for f ,cols in col_sets.items():
    miss=all_cols-cols
    if miss:
        for c in sorted(miss):
            print(f"- {c}")
        print("---------")

```
