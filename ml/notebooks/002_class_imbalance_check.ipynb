```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("../data/processed/train.csv")
val=pd.read_csv("../data/processed/validation.csv")
test=pd.read_csv("../data/processed/test.csv")
```

```python
for name,df in [("train",train),("val",val),("test",test)]:
    print(f"{name} - {len(df)} rows")
    print(df["Label"].value_counts())
    print(df["Label"].value_counts(normalize=True).round(4))
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,4))
for ax,(name,df) in zip(axes,[("train",train),("val",val),("test",test)]):
    df["Label"].value_counts().plot(kind="bar",ax=ax,title=name)
    ax.set_xlabel("Label")
    ax.set_ylabel("Count")

plt.tight_layout()
plt.show()
```
